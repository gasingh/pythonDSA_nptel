```python
# PROBLEM 2

def repfree(sampleStr):
    """ finds if there are internal repeat characters in an input string
    03:51 09/02/2022
    """
    strLst = sampleStr[:]
    repetitionBln = False
    for i,j in enumerate(strLst):
        foundCnt = 0
        for m,n in enumerate(strLst):
            if j == n:
                foundCnt = foundCnt + 1
        if foundCnt >= 2:
            repetitionBln = True
        # print i,j, " | ", foundCnt

    if repetitionBln:
        # some repetition was found in the string
        return False
    else:
        # no repetitions in the input string
        return True

print(repfree("tester"))
print(repfree("zb%78"))
print(repfree("(7)(a"))
print(repfree("a)*(?"))
print(repfree("abracadabra"))
```
